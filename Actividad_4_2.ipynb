```python
#Importamos librer√≠as requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_csv('Oporto_Portugal.csv')
df
```

```python
df.info()
```

```python
#A) Los registros de 5 ‚Äúhost name‚Äù diferentes
filtroa = df["host_name"].dropna().unique()[:5]
print(f"nombre de los cinco hosts diferentes:{filtroa}")
```

```python
#B) Los registros de los host que se hallan unido a Airbnb ‚Äúdespues del a√±o 2020‚Äù
df["host_since"] = pd.to_datetime(df["host_since"])
filtrob = df[df["host_since"] > "2020-12-31"]
filtrob[["host_name", "host_since"]].head(3)
```

```python
#C) Los registros de los host que responden ‚Äúa m√°s tardar en 1 d√≠a‚Äù
filtroc = df[df["host_response_time"] == "within a day"]
filtroc[["host_name", "host_response_time"]].head(3)
```

```python
#D) Los registros de las filas ‚ÄúM√∫ltiplos de 200‚Ä¶ (400, 600, 800, etc.)
filtrod = df.iloc[0::200]
filtrod
```

```python
#E) Los registros de los tipos de cuarto: ‚ÄúEntire home‚Äù y ‚ÄúPrivate room
filtroe = df[df["room_type"].isin(["Entire home/apt", "Private room"])]
filtroe.head()
```

```python
#F) Los registros que ‚Äúa lo mucho cuenten con 4 camas
filtrof = df[df["beds"] <= 4]
filtrof.head()
```

```python
#G) Los registros que ‚Äúsean superhost‚Äù
filtrog = df[df["host_is_superhost"] == "t"]
filtrog.head()
```

```python
#H) Los registros de las columnas ‚Äúimpares‚Äù
filtroh = df.iloc[:, ::2]  
filtroh.head()
```

```python
filtroa_df = pd.DataFrame(filtroa, columns=["host_name"])
filtroa_df.to_csv("FiltroA.csv", index=False)
```
