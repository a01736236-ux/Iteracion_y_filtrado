```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_csv('Oporto_Portugal.csv')
df
```

```python
df.info()
```

```python
#A) Los registros de 5 “host name” diferentes
filtroa = df["host_name"].dropna().unique()[:5]
print(f"nombre de los cinco hosts diferentes:{filtroa}")
```

```python
#B) Los registros de los host que se hallan unido a Airbnb “despues del año 2020”
df["host_since"] = pd.to_datetime(df["host_since"])
filtrob = df[df["host_since"] > "2020-12-31"]
filtrob[["host_name", "host_since"]].head(3)
```

```python
#C) Los registros de los host que responden “a más tardar en 1 día”
filtroc = df[df["host_response_time"] == "within a day"]
filtroc[["host_name", "host_response_time"]].head(3)
```

```python
#D) Los registros de las filas “Múltiplos de 200… (400, 600, 800, etc.)
filtrod = df.iloc[0::200]
filtrod
```

```python
#E) Los registros de los tipos de cuarto: “Entire home” y “Private room
filtroe = df[df["room_type"].isin(["Entire home/apt", "Private room"])]
filtroe.head()
```

```python
#F) Los registros que “a lo mucho cuenten con 4 camas
filtrof = df[df["beds"] <= 4]
filtrof.head()
```

```python
#G) Los registros que “sean superhost”
filtrog = df[df["host_is_superhost"] == "t"]
filtrog.head()
```

```python
#H) Los registros de las columnas “impares”
filtroh = df.iloc[:, ::2]  
filtroh.head()
```

```python
filtroa_df = pd.DataFrame(filtroa, columns=["host_name"])
filtroa_df.to_csv("FiltroA.csv", index=False)
```
